```python
for station in frame.station.unique():
    station_df=frame[frame.station == station]
    station_df.set_index(["No", "Date"], inplace=True)
    station_df.drop("station", axis=1, inplace=True)
    station_df.to_csv("station_{}.csv".format(station))
```

```python
df_from_each_file = (pd.read_csv("air/{}".format(f)) for f in os.listdir("air"))

#list_cols=[ col for col in list(df_from_each_file.columns) if col not in ["station", "Date"]]
frame = pd.concat(df_from_each_file, ignore_index=True)
frame['Date'] = pd.to_datetime(frame[['year', 'month', 'day', "hour"]])
frame.drop(["year", "month", "day", "hour"], inplace=True, axis=1)
frame.fillna(method="backfill", inplace=True)
frame = pd.get_dummies(frame, columns=['wd'])
frame.set_index([ "station", "Date"], inplace=True)
frame = frame.apply(pd.to_numeric, errors='coerce')
```
